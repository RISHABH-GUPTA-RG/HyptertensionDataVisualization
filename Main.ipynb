We will convert Object data (Strings) to Numeric format to see correlations and perform statistacal calculation on it

In this we will Convert:

>Gender:
>>  Male into 0<br>
>>  Female into 1<br>
>>  <br>

>work_type:
>>  children into 0<br>
>>  Govt_job into 1<br>
>>  Never_worked into 2<br>
>>  Private into 3<br>
>>  Self_employed into 4<br>
>>  <br>

>ever_married:
>>  No into 0<br>
>>  Yes into 1<br>
>>  <br>

>Residence_type:
>>  Rural into 0<br>
>>  Urban into 1<br>
>>  <br>


```python
df.replace({'gender': {'Male': 0, 'Female': 1}}, inplace=True)
df.replace({'work_type': {'children': 0, 'Govt_job': 1,
           'Never_worked': 2, 'Private': 3, 'Self-employed': 4}}, inplace=True)
df.replace({'Residence_type': {'Rural': 0, 'Urban': 1}}, inplace=True)
df.replace({'ever_married': {'No': 0, 'Yes': 1}}, inplace=True)
df.drop(columns=['id', 'smoking_status'], axis=1, inplace=True)
df
```

There are excessive outliers so we are gonna remove them

```python
# Identify the quartiles
q1,q3 =np.percentile(df['avg_glucose_level'],[0,90]) #type:ignore
# Calculate the interquartile range
iqr = q3 - q1
# Calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
# Drop the outliers
df = df[(df['avg_glucose_level'] >= lower_bound) 
                & (df['avg_glucose_level'] <= upper_bound)]

# Identify the quartiles
q1,q3 =np.percentile(df['bmi'],[0,90]) #type:ignore
# Calculate the interquartile range
iqr = q3 - q1
# Calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
# Drop the outliers
df = df[(df['bmi'] >= lower_bound) 
                & (df['bmi'] <= upper_bound)]
```

Note:-
 The kde shows the density of the feature for each value of the target. 
 There are usually 2 colored humps representing the 2 values of TARGET.
 If the humps are well-separated and non-overlapping, then there is a correlation with the TARGET.
 If the humps are overlapping a lot, then that means the feature is not well-correlated with the TARGET,
 because the TARGET is equally as common on those values of the feature.

#### Seperating the data and labels

### Data standardisation

```python
# Before standardisation
X_data[:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_data)
X = scaler.transform(X_data)
```

### SPLIT DATA IN TEST AND TRAIN FOR HYPERTENSION PREDICTION

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_hypertension, test_size = 0.2)
```

### Importing Libraties for Training ML models


```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.naive_bayes import GaussianNB
```

### Training Diffrent ML Models

```python
modelLR= LogisticRegression()
modelDT = DecisionTreeClassifier(random_state=0)
modelSVC = SVC(kernel='linear')
modelSVCrbf = SVC(kernel='rbf',C=1)
modelET=ExtraTreesClassifier()
modelKnn = KNeighborsClassifier(n_neighbors=6)
modelRFC = RandomForestClassifier(n_estimators=22)
modelNB = GaussianNB()
```

```python
modelLR.fit(X_train,Y_train)
modelDT.fit(X_train,Y_train)
modelSVC.fit(X_train, Y_train)
modelSVCrbf.fit(X_train, Y_train)
modelKnn.fit(X_train, Y_train)
modelRFC.fit(X_train, Y_train)
modelNB.fit(X_train, Y_train)
modelET.fit(X_train,Y_train)
```

### Finding the accuracy score on test dataset

```python
accLR=modelLR.score(X_test,Y_test)
accDT=modelDT.score(X_test,Y_test)
accSVC=modelSVC.score(X_test,Y_test)
accSVCrbf=modelSVCrbf.score(X_test,Y_test)
accKnn=modelKnn.score(X_test,Y_test)
accRFC=modelRFC.score(X_test,Y_test)
accNB=modelNB.score(X_test,Y_test)
accET=modelET.score(X_test,Y_test)
```

```python
plt.figure(figsize=(20,6))
fig=plt.bar(['Decision Tree', 'Logistic', 'SVM linear', 'SVM rbf', 'KNeighborsClassifier',
        'RandomForestClassifier', 'Naive Bayes','ExtraTreesClassifier'], [accDT, accLR, accSVC, accSVCrbf, accKnn, accRFC, accNB,accET]) #type:ignore

plt.xlabel("Algorithms")
plt.ylabel("Accuracy")
plt.show()
print(accDT,accLR,accSVC,accSVCrbf,accKnn,accRFC,accNB,accET)
```

### ROC curve

```python
from sklearn.metrics import roc_curve,auc
```

```python
roc_data={
    'True':Y_test,
    'Logistics':modelLR.predict(X_test),
    'Decision Tree':modelDT.predict(X_test),
    'SVM linear':modelSVC.predict(X_test),
    'SVM Rbf':modelSVCrbf.predict(X_test),
    'KNN':modelKnn.predict(X_test),
    'Random Forest':modelRFC.predict(X_test),
    'Extra tree':modelET.predict(X_test),
    'Naive Bayes':modelNB.predict(X_test)
}
```

```python
# Plot ROC curve for each model
plt.figure(figsize=(7, 5))
 
for model in ['Logistics', 'Decision Tree', 'SVM linear', 'SVM Rbf', 'KNN', 'Random Forest', 'Extra tree', 'Naive Bayes']:
    fpr, tpr, _ = roc_curve(roc_data['True'], roc_data[model])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{model} (AUC = {roc_auc:.2f})')
 
# Plot random guess line
plt.plot([0, 1], [0, 1], 'r--', label='Random Guess')
 
# Set labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Two Models')
plt.legend()
plt.show()
```
